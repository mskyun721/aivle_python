### merge()
 - merge(df1, df2, how=, on=)
    - how : left, right, inner, outer [ default inner ]
    - on : merge key column

```python
sales = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/sales.csv")
products = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/products.csv")
```

```python
sales
```

```python
products
```

```python
tmp_merg_in = pd.merge(sales, products, how='inner', on='ProductID')
tmp_merg_out = pd.merge(sales, products, how='outer', on='ProductID')
tmp_merg_lf = pd.merge(sales, products, how='left', on='ProductID')
tmp_merg_ri = pd.merge(sales, products, how='right', on='ProductID')
```

     - axis : default 0 ( 0 - 행 / 1 - 열)

# 데이터 프레임 추가기능

### Rolling
 - 시계열 데이터에 대해 이동 집계 계산
 - rolling(day, min_periods=)
    - day : int, 이동평균 구할 값
    - min_periods : 최소 데이터 값 

```python
stock = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/SK.csv') 
stock = stock[['Date', 'Close', 'Volume']]
stock.head()
```

```python
stock['Close_3'] = stock['Close'].rolling(3).mean()
stock
```

```python
stock['Close_3mean'] = stock['Close'].rolling(3, min_periods=1).mean()
stock['Close_3max'] = stock['Close'].rolling(3, min_periods=1).max()
stock
```

### Shift()
 - 데이터 행 이동
 - default 1

```python
stock['Close_shift'] = stock['Close'].shift()
stock['Close_shift2'] = stock['Close'].shift(2)
stock
```

```python
### cvs 파일 불러오기 : read_csv()
# 옵션
# usecols : 불러올 칼럼 지정(칼럼이름을 알고 있어야 함)
# sep: 구분자 지정(기본값 = 콤마)
# header: 헤더가 될 행 번호 지정(기본값 = 0)
# index_col: 인덱스 열 지정(기본값 = False)
# names: 열 이름으로 사용할 문자열 리스트
# skiprows: 처음 몇 줄을 무시할 것인지 지정, 리스트로 지정 가능

# 파일을 불러올 경우 파일 경로 지정

path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/titanic_simple.csv'
df = pd.read_csv(path)
df
```

# 데이터 프레임 탐색

### 기본 데이터 정보
- head(): 상위 데이터 확인. default 5
- tail(): 하위 데이터 확인. default 5
- shape: 데이터프레임 크기
- index: 인덱스 정보 확인
- values: 값 정보 확인
- columns: 열 정보 확인
- dtypes: 열 자료형 확인
- info(): 열에 대한 상세한 정보 확인
- describe(): 기초통계정보 확인

```python
print(df.head())
print()
print(df.tail())
print()
print(df.shape)
print()
print(df.index)
print()
print(df.values)
print()
print(df.columns)
print()
print(df.dtypes)
print()
print(df.info())
print()
print(df.describe())
```

### 데이터 정렬
 - sort_values(by, axis, ascending, inplace, kind, na_position, ignore_index)
    - by : 기준 column
    - axis : 0 - 열 방향 정렬 / 1 - 행 방향 정렬. default 0
    - ascending : [True, False] 오름차순. default True
    - inplace : [True, False] 정렬 데이터 저장 유므. default False

```python
df.sort_values(by='Age')

df.sort_values(by=['Survived', 'Age'], ascending=[True, False], ignore_index=True)
```

### 데이터 프레임 집계
 - unique() : 컬럼의 고유 값
 - value_counts() : 고유 값의 개수
 - sum() / max() / mean() / min() / median()
 - groupby()

```python
df['Sex'].unique()
```

```python
df['Sex'].value_counts()
```

```python
# groupby() : by='열' 별로 집계
df.columns
df.groupby(by=['Pclass','Sex'], as_index=True)['Age'].mean() # index 없음
df.groupby(by=['Pclass','Sex'], as_index=False)['Age'].mean() # index 있음

df.groupby(by=['Pclass','Sex'], as_index=True)[['Age','Fare']].mean() # index 없음
df.groupby(by=['Pclass','Sex'], as_index=False)[['Age','Fare']].mean() # index 있음
```

# 데이터 프레임 조회

### 특정 열 조회

```python
df[['Sex','Age']]
df.Age
```

DataFrame.loc[ 조건 ]

```python
df.loc[df['Age'] >= 20]
```

```python
# isin(list) : list 포함 데이터
df.loc[df['Embarked'].isin(['Southhampton', 'Cherbourg'])]
```

```python
# between(st, ed) : st ~ ed 까지 범위
df.loc[df['Age'].between(0,20)].sort_values(by='Age', ignore_index=True)
```

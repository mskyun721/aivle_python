### Plot 차트
* plt.plot(1차원 값)
    * x축 : 인덱스
    * y축 : 1차원 값
    * 라인차트를 그려줍니다.

```python
# Date = to_datetime
print(air.dtypes)
air['Date'] = pd.to_datetime(air['Date'])
```

```python
# plt.plot(x, y, style, data, label)

plt.plot('Date', 'Ozone', '', data=air, label='Ozone')
plt.plot('Date', 'Temp', '', data=air, label='Temp')
plt.plot('Date', 'Wind', '', data=air, label='Wind')
plt.plot('Date', 'Solar.R', '', data=air, label='Solar.R')

# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Figure')
# title
plt.title('Air')
# 사이즈 (가로, 세로)
plt.figure(figsize=(15,5))

plt.show()
```

```python
# 차트 분리 : plt.Subplot()
plt.subplot(2,2,1)
plt.plot('Date', 'Ozone', 'bs-', data=air, label='Ozone')
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Ozone')
# title
plt.title('Air')



plt.subplot(2,2,2)
plt.plot('Date', 'Temp', 'go--', data=air, label='Temp')
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Temp')
# title
plt.title('Air')


plt.subplot(2,2,3)
plt.plot('Date', 'Wind', 'rx:', data=air, label='Wind')
# 격자
plt.grid()
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Wind')
# title
plt.title('Air')



plt.subplot(2,2,4)
plt.plot('Date', 'Solar.R', 'cp-.', data=air, label='Solar.R')
# 격자
plt.grid()
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Solar.R')
# title
plt.title('Air')


# 차트 간격
plt.tight_layout()
plt.show()
```

##### 추가 기능

```python
# 1. 축 범위 : xlim / ylim (st, ed)
dict1 = {'v1': [1,2,3,4,5], 'v2': [2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.plot('v1', 'v3', 'rs-', data = dict1, label = 'mango')
plt.xlim(0, 6)
plt.ylim(0, 8)
plt.grid()
plt.show()
```

```python
# 수평 / 수직선 : axhline / axvline (축 위치, color, linestyle)
dict1 = {'v1': [1,2,3,4,5], 'v2': [2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.plot('v1', 'v3', 'rs-', data = dict1, label = 'mango')
plt.plot('v1', 'v2', 'bs-', data = dict1, label = 'apple')
plt.axhline(4.3, color = 'grey', linestyle = '--')
plt.axvline(2.4, color = 'grey', linestyle = '--')
plt.show()
```

```python
# 그래프 텍스트 : text(x축, y축, 텍스트)
dict1 = {'v1': [1,2,3,4,5], 'v2': [2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.plot('v1', 'v3', 'rs-', data = dict1, label = 'mango')
plt.axhline(5.3, color = 'grey', linestyle = '--')
plt.axvline(2.4, color = 'grey', linestyle = '--')

plt.text(2.5, 4.2, '2.4')
plt.text(1.2, 5.4, '5.3')

plt.show()
```

* 산술평균

```python
### 산술평균
a = [1, 2, 3, 3, 4, 4, 4, 5, 6, 7, 8, 9, 10, 11, 15, 17, 17, 19, 20, 25, 30]
print(np.mean(a))
print(ti['Fare'].mean())
```

* 조화평균
    * a, b : [[두 수(a,b)의 역수]의 평균]의 역수
    * 분자가 같고, 분모가 다른 두 숫자에 대한 평균 계산시 사용 (추후 모델 평가시 중요하게 사용됩니다. 여기서는 가볍게~~)
    * 1) 두 수의 역수의 평균 :
$\frac{(\frac{1}{a}+\frac{1}{b})}{2}$
    * 2) 1)의 역수 :
$\frac{2*a*b}{(a+b)}$

```python
### 조화평균
# 거리 20 km
# 갈때 속력 60 km
# 올때 속력 80 km
# 평균 속력은?

a1, a2 = 60, 80
hm = 2*a1*a2 / (a1 + a2)
print(hm)
```

### heatmap : 두 범주의 집계 시각화


```python
ti.isna().sum()
tmp = ti.groupby(['Survived', 'Pclass'], as_index=False)['PassengerId'].count()
ti_pivot = tmp.pivot('Survived', 'Pclass', 'PassengerId')
seaborn.heatmap(ti_pivot
                , annot=True # 수치 표현
                , fmt='d' # 수치 정수
                , linewidths=0.2) # 간격 설정
plt.show()
```

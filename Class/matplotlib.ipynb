```python
# matplotlib import
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
import numpy as np
```

```python
# csv
ti = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/titanic_simple.csv')
air = pd.read_csv('https://bit.ly/3qmthqZ')
```

* 데이터 시각화

> 수많은 데이터를 한눈에 파악할 수 있는 방법  
① 시각화 도구 : histogram, boxplot, densityplot, scatter, ...  
③ 수치화 도구 : min, max, mean, std, 사분위수, ...

* Focus
    * 우리가 다루는 데이터는 비즈니스를 담고 있습니다.
    * 데이터 시각화의 목적은 
    * 아름다운 그래프가 아니라, 
    * 통계적인 해석을 넘어,
    * 한눈에 파악해야 할 정보는 비즈니스의 인사이트 입니다.

* 한계
    * 그래프와 통계량은 대부분 정보(원본)를 요약합니다.
    * 요약을 하는 관점에 따라 결과가 달라집니다.
    * 어떤 식으로든 요약을 하면, 정보의 손실이 발생 됩니다. 

### Plot 차트
* plt.plot(1차원 값)
    * x축 : 인덱스
    * y축 : 1차원 값
    * 라인차트를 그려줍니다.

```python
# Date = to_datetime
print(air.dtypes)
air['Date'] = pd.to_datetime(air['Date'])
```

```python
# plt.plot(x, y, style, data, label)

plt.plot('Date', 'Ozone', '', data=air, label='Ozone')
plt.plot('Date', 'Temp', '', data=air, label='Temp')
plt.plot('Date', 'Wind', '', data=air, label='Wind')
plt.plot('Date', 'Solar.R', '', data=air, label='Solar.R')

# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Figure')
# title
plt.title('Air')
# 사이즈 (가로, 세로)
plt.figure(figsize=(15,5))

plt.show()
```

```python
# 차트 분리 : plt.Subplot()
plt.subplot(2,2,1)
plt.plot('Date', 'Ozone', '', data=air, label='Ozone')
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Ozone')
# title
plt.title('Air')

plt.subplot(2,2,2)
plt.plot('Date', 'Temp', '', data=air, label='Temp')
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Temp')
# title
plt.title('Air')

plt.subplot(2,2,3)
plt.plot('Date', 'Wind', '', data=air, label='Wind')
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Wind')
# title
plt.title('Air')

plt.subplot(2,2,4)
plt.plot('Date', 'Solar.R', '', data=air, label='Solar.R')
# 범례
plt.legend()
# x 축
plt.xlabel('Date')
# yc 축
plt.ylabel('Solar.R')
# title
plt.title('Air')

# 차트 간격
plt.tight_layout()

plt.show()
```
